```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Sample - Superstore.csv",encoding = "latin-1")
data.head()
```

```python
data.info()
```

# Change Datatype of (Order Date,Ship Date)

```python
data["Order Date"]=pd.to_datetime(data["Order Date"])
data["Ship Date"]=pd.to_datetime(data["Ship Date"])
```

```python
data.info()
```

```python
data.columns=data.columns.str.replace(" ","_")
```

```python
data.columns
```

```python
data.describe()
```

# Extract Month & Year

```python
data["Year"]=data["Order_Date"].dt.year
```

```python
data["Month"]=data["Order_Date"].dt.month_name()
```

```python
data.head()
```

# Monthly Sales Analysis

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
Monthly_Sales=data.groupby("Month")["Sales"].sum().reset_index().sort_values(by="Sales",ascending = False)
Monthly_Sales
```

```python
plt.figure(figsize=(15,5))

ax=sns.lineplot(x="Month", y="Sales", data=Monthly_Sales)
plt.title("Monthly Sales",fontsize=20)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

# Category Wise Sales Analysis

```python
Category_sales=data.groupby("Category")["Sales"].sum().reset_index().sort_values(by="Sales",ascending = False)
Category_sales
```

```python
plt.figure(figsize=(6,4))
ax=sns.barplot(y="Sales",x="Category",data=Category_sales)
plt.title("Category_sales",fontsize=18)

for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}')
plt.show()
```

# Sales And Profit By Segment

```python
analysis=data.groupby("Segment")[["Sales","Profit"]].sum().reset_index()
analysis
```

```python
analysis_melt = analysis.melt(
    id_vars="Segment",
    value_vars=["Sales", "Profit"],
    var_name="Metric",
    value_name="Value"
)

ax = sns.barplot(
    x="Segment",
    y="Value",
    hue="Metric",
    data=analysis_melt
)

# add data labels
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}')

plt.title("Sales and Profit by Customer Segment", fontsize=16)
plt.xlabel("Customer Segment")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# Sales to Profit Ratio

```python
sales_profit_ratio = (
    data.groupby("Segment", as_index=False)
        .agg(
            Total_Sales=("Sales", "sum"),
            Total_Profit=("Profit", "sum")
        )
)

sales_profit_ratio["Sales_to_Profit_Ratio"] = (
    sales_profit_ratio["Total_Sales"] / sales_profit_ratio["Total_Profit"]
).round(2)
sales_profit_ratio
```
